```python
import torch
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np
import random
import os
import glob
from PIL import Image
```

```python
%matplotlib inline
def show(img, filename='img.pdf',dpi=600):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1,2,0)), interpolation=None)
    plt.axis('off') 
    plt.savefig(filename,pad_inches=0.0,bbox_inches='tight',dpi=dpi)
```

# CIFAR-10

```python
imgs = glob.glob('/home/datasets/Flythings3D/MEGAN/exps/arch_train_cifar10_2020_12_22_13_39_26/Samples/fid_buffer/*')
imgs = np.array(imgs)
imgs.shape
```

```python
imglist = imgs[random.sample(range(1000,50000),64)]
imglist = [transforms.ToTensor()(Image.open(x)) for x in imglist]
# print(imglist[0].shape)
img = make_grid(imglist, nrow=8,scale_each=True)
show(img, 'cifar10_origin.pdf')
# height=width=10
# plt.figure(figsize=[height, width])
# plt.axis('off') 
# # plt.tight_layout()
# plt.imshow(transforms.ToPILImage()(img))
# plt.savefig('cifar10.pdf')
```

```python
height=width=10
plt.figure(figsize=[height, width])
plt.imshow(transforms.ToPILImage()(img))
plt.savefig('cifar10.pdf')
```

# STL10

```python
imgs = glob.glob('exps/arch_test_stl10_2020_12_30_16_17_28/Samples/fid_buffer/*')
imgs = np.array(imgs)
imgs.shape
```
